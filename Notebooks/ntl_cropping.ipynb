```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.plot import show

# Open the GeoTIFF file
geo_tiff_path = r'C:\Users\rishi\ml_projects\climate\NTL cropped and named\20180601-20180630avg_rade9h.tif'  # Replace with your file path
#C:\Users\rishi\ml_projects\climate\SVDNB_npp_20190101-20190131_00N060E_vcmslcfg_v10_c201905201300.cf_cvg.tif
with rasterio.open(geo_tiff_path) as src:
    # Read the first band (MODIS data can have multiple bands)
    band1 = src.read(1)
    # Get metadata
    profile = src.profile
    print("GeoTIFF Metadata:")
    print(profile)
    bbox=src.bounds
    print(bbox)
    # Get the affine transformation
    transform = src.transform
    
    # Calculate the extent of the image for plotting
    # Extent is defined as (xmin, xmax, ymin, ymax) in the image's coordinate system
    xmin, ymax = transform * (0, 0)
    xmax, ymin = transform * (band1.shape[1], band1.shape[0])
    extent = [xmin, xmax, ymin, ymax]
    
    # Plot the data
    plt.figure(figsize=(8, 6))
    plt.imshow(band1, cmap='gray', extent=extent)  # Use the 'extent' to set the correct coordinates
    plt.title('MODIS NDVI')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.colorbar(label='NDVI')
    plt.show()
```

```python
from rasterio.enums import Resampling

geo_tiff_path = r'C:\Users\rishi\ml_projects\climate\only rade\20180601-20180630avg_rade9h.tif'
height_factor=1/7.766
width_factor=1/15.6875
with rasterio.open(geo_tiff_path) as dataset:
    data = dataset.read(
    out_shape=(
        dataset.count,
        int(dataset.height * height_factor),
        int(dataset.width * width_factor)
    ),
    resampling=Resampling.average
)

# Scale image transform
transform = dataset.transform * dataset.transform.scale(
    (dataset.width / data.shape[-1]),
    (dataset.height / data.shape[-2])
)

# Define the new metadata for the output file
new_meta = dataset.meta.copy()
new_meta.update({
    'height': data.shape[1],
    'width': data.shape[2],
    'transform': transform
})

# Save the resampled image
with rasterio.open('resampled_image_nlt.tif', 'w', **new_meta) as dst:
    dst.write(data)

new_path='resampled_image_nlt.tif'
with rasterio.open(new_path) as dataset:
    band1=dataset.read()
```

```python
import rasterio
from rasterio.enums import Resampling
import os

dir_path=r"C:\Users\rishi\ml_projects\climate\new_rade"
output_dir=r"C:\Users\rishi\ml_projects\climate\resampled_new"
for file in os.listdir(dir_path):
    geo_tiff_path = os.path.join(dir_path,file)
    output_path=os.path.join(output_dir,file)
    height_factor=1/7.766
    width_factor=1/15.6875
    with rasterio.open(geo_tiff_path) as dataset:
        data = dataset.read(
        out_shape=(
            dataset.count,
            int(dataset.height * height_factor),
            int(dataset.width * width_factor)
        ),
        resampling=Resampling.average
    )

    # Scale image transform
    transform = dataset.transform * dataset.transform.scale(
        (dataset.width / data.shape[-1]),
        (dataset.height / data.shape[-2])
    )

    # Define the new metadata for the output file
    new_meta = dataset.meta.copy()
    new_meta.update({
        'height': data.shape[1],
        'width': data.shape[2],
        'transform': transform
    })

    # Save the resampled image
    with rasterio.open(output_path, 'w', **new_meta) as dst:
        dst.write(data)
```

```python
import rasterio
import numpy as np
import os
from datetime import datetime

# Directory containing the TIFF files
dir_path = r"C:\Users\rishi\ml_projects\climate\resampled_new"

# Directory where the CSV files will be saved
output_base_dir = r"C:\Users\rishi\ml_projects\climate\Intermediate outputs\NTL_outputs"

# Iterate over each file in the directory
for file in os.listdir(dir_path):
    if file.endswith(".tif"):  # Process only .tif files
        with rasterio.open(os.path.join(dir_path, file)) as dataset:
            # Read the first band
            band1 = dataset.read(1)
            
            # Extract the month and year from the filename (assuming file naming convention)
            # Example filename: 20180701-20180731avg_rade9h.tif
            date_str = file[:8]  # Extracting the '20180701' part
            date_obj = datetime.strptime(date_str, "%Y%m%d")  # Convert to datetime object
            month_name = date_obj.strftime("%B")  # Get full month name (e.g., "July")
            year = date_obj.strftime("%Y")  # Get the year (e.g., "2018")
            
            # Create the output directory based on the year and month
            month_dir = os.path.join(output_base_dir, year)
            os.makedirs(month_dir, exist_ok=True)
            
            # Construct the output file path
            output_filename = f"{month_name}_NTL.csv"
            save_path = os.path.join(month_dir, output_filename)
            
            # Save the band data as a CSV
            np.savetxt(save_path, band1, delimiter=',')
```

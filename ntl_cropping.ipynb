```python
import os
from osgeo import gdal

dir_path=r"D:\NTL data - Copy"
target_bbox = [78.00405826, 16.93264351, 79.04971836, 17.90150706]
new_dir=r"C:\Users\rishi\ml_projects\climate\ntl_cropped"
files=os.listdir(dir_path)

for file in files:
    output_image = os.path.join(new_dir,file)
    input_image= os.path.join(dir_path,file)
    # Open the image using GDAL
    dataset = gdal.Open(input_image)

    # Crop the image using gdal.Warp
    gdal.Warp(output_image, dataset, outputBounds=target_bbox)

    # Close the dataset
    dataset = None
```

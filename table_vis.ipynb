```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

urban_df=pd.read_csv(r"C:\Users\rishi\ml_projects\climate\Urban_Areas_Regression_Results.csv")
periurban_df=pd.read_csv(r"C:\Users\rishi\ml_projects\climate\Peri-Urban_Areas_Regression_Results.csv")
```

```python
pollutants = urban_df["Air Pollutants"]


urban_values = urban_df.drop(columns=["Air Pollutants"]).values
periurban_values = periurban_df.drop(columns=["Air Pollutants"]).values

bar_width = 0.15
x = np.arange(len(pollutants))

# Colors for differentiation
colors = ['b', 'r', 'g']  # Blue for NTL, Green for NLST, Red for NDVI
colors2= ['cyan','darkorange','lime']

plt.figure(figsize=(10,6))

for i in range(3):
    plt.bar(x + (i - 1) * bar_width, urban_values[:, i], bar_width, label=f"Urban {['NTL', 'NLST', 'NDVI'][i]}", alpha=0.7, color=colors[i],edgecolor='black')
    plt.bar(x + (i - 1) * bar_width, periurban_values[:, i], bar_width, label=f"Peri-Urban {['NTL', 'NLST', 'NDVI'][i]}", alpha=0.7,color=colors2[i], edgecolor='black')


plt.xticks(ticks=x, labels=pollutants, ha="right")
plt.xlabel("Air Pollutants")
plt.ylabel("Regression Coefficients")
plt.title("Comparison of Urban and Peri-Urban Areas for NTL, NLST, and NDVI")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
pollutants = urban_df["Air Pollutants"]

urban_values = urban_df.drop(columns=["Air Pollutants"]).values
periurban_values = periurban_df.drop(columns=["Air Pollutants"]).values

bar_width = 0.12
x = np.arange(len(pollutants))  # One group per pollutant

# Colors: alternating Urban and Peri-Urban
urban_colors = ['b', 'r', 'g']         # Urban: NTL, NLST, NDVI
periurban_colors = ['cyan', 'tomato', 'lime']  # Peri-Urban: NTL, NLST, NDVI

plt.figure(figsize=(10,6))

# Plot each pair of Urban and Peri-Urban bars next to each other
for i in range(3):  # 0: NTL, 1: NLST, 2: NDVI
    plt.bar(x + (2 * i - 2.5) * bar_width, urban_values[:, i], bar_width,
            label=f"Urban {['NTL', 'NLST', 'NDVI'][i]}", color=urban_colors[i], edgecolor='black')
    plt.bar(x + (2 * i - 1.5) * bar_width, periurban_values[:, i], bar_width,
            label=f"Peri-Urban {['NTL', 'NLST', 'NDVI'][i]}", color=periurban_colors[i], edgecolor='black')

plt.xticks(ticks=x, labels=pollutants, ha="right")
plt.xlabel("Air Pollutants")
plt.ylabel("Regression Coefficients")
plt.title("Comparison of Urban and Peri-Urban Areas for NTL, NLST, and NDVI")
plt.legend(loc='upper right', ncol=1)
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```
